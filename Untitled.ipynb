```python
# Convert with error handling
df['Join Date'] = pd.to_datetime(df['Join Date'], errors='coerce' ,format='%Y-%m-%d')
```

```python
df.columns
```

```python
df["Join Date"]
```

```python
year_data = df[df['Join Date'].dt.year == 2023]
```

```python
year_data.groupby(year_data['Join Date'].dt.month)['Monthly Revenue'].sum().plot(kind='bar', color='teal')
plt.xticks(ticks=range(0, 7), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul'])
plt.xlabel('Month')
plt.ylabel('Total Monthly Revenue')
plt.title(f'Total Monthly Revenue for 2023')
```

```python
year_data = df[df['Join Date'].dt.year == 2022]
year_data.groupby(year_data['Join Date'].dt.month)['Monthly Revenue'].sum().plot(kind='bar', color='navy')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.xlabel('Month')
plt.ylabel('Total Monthly Revenue')
plt.title(f'Total Monthly Revenue for 2022')
```

```python
# Filter data for June
june_data = df[df['Join Date'].dt.month == 10]

# Group by region and sum the revenue for June
june_data.groupby('Country')['Monthly Revenue'].sum().plot(kind='bar', color='blueviolet', figsize=(8, 6))
plt.xlabel('Region')
plt.ylabel('Total Monthly Revenue')
plt.title('Total Monthly Revenue in October per Region')
```
